```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv("../data/d10800_BS2_JoinedTrans.csv")
df.head()
```

```python
X = df.drop(columns=['src_file', 'row_id'])
X.head()
```

```python
X = X.apply(pd.to_numeric, downcast='integer')
```

```python
X.shape
```

```python
y = df['src_file']
```

```python
from sklearn.feature_selection import VarianceThreshold
from scipy.sparse import csr_matrix
```

```python
X_sparse = csr_matrix(X)
X_sparse.shape
```

```python
selector = VarianceThreshold(threshold=0.05)
X_selected = selector.fit_transform(X_sparse)
X_selected
```

```python
X = pd.DataFrame(X_selected.todense())
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
string_to_int_mapping = {string: index for index, string in enumerate(np.unique(y))}

y_mapped = np.array([string_to_int_mapping[string] for string in y])

string_to_int_mapping
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_mapped, test_size=0.3, stratify=y, random_state=0)
```

```python
from joblib import dump
dump((X_train, X_test, y_train, y_test), 'preprocessed_data_BS2.joblib')
```

```python
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

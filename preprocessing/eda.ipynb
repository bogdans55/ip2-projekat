```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/preprocessed_data_BS1.csv')
```

```python
X = data.drop(['Group'], axis=1)
y = data['Group']
```

```python
flattened_series = X.apply(pd.Series.value_counts).sum(axis=1)
```

```python
flattened_df = flattened_series.reset_index()
flattened_df.columns = ['Value', 'Count']

plt.figure(figsize=(10, 6))
sns.barplot(x='Value', y='Count', data=flattened_df, palette='viridis')

plt.title('Broj pojavljivanja vrednosti u skupu podataka BS1')
plt.xlabel('Vrednosti')
plt.ylabel('Broj pojavljivanja')
plt.xticks(rotation=45)
plt.show()
```

```python
y_value_counts = y.value_counts()

plt.figure(figsize=(10, 6))
plt.pie(y_value_counts, labels=y_value_counts.index, autopct='%1.1f%%', startangle=140)

plt.title('Udeo redova svake klase u skupu podataka')
plt.axis('equal')

plt.show()
```
